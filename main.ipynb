```python
import os
import math
import pandas as pd
import matplotlib.pyplot as plt
import dataframe_image as dfi
import numpy as np
from pypdf import PdfMerger
import json
import glob
from fpdf import FPDF 
import copy
from matplotlib.backends.backend_pdf import PdfPages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import copy
import networkx as nx

########################################
import telebot
import pandas as pd
from telebot import types
import datetime 
import modir
import json

chat_ids = {}
code_to_chat_dict = {}

TOKEN = 'YOUR TELEGRAM BOT TOKEN
bot = telebot.TeleBot(TOKEN) #create the BOTTTTT
########################################


#plt.style.use('default')  
#plt.rcParams['axes.facecolor'] = 'black' #"azure"#'black'#"azure"#'khaki'#'lavender'
    
def int_to_str (list):
    for i in range (len(list)):
        list[i]=str(list[i])
    return list

def str_to_int (list):
    for i in range (len(list)):
        list[i]=int(list[i])
    return list

def history(StName):
    hist_list = main_dict[StName][4]
    participant = main_dict[StName][0]
    if len(hist_list)==0:
        print(participant, ' did not speak. ')
    for i in range(len(hist_list)):
        print(participant, ' spoke for ', data['messages'][hist_list[i]]["duration_seconds"],' seconds on ', data['messages'][hist_list[i]]["date"][0:-9] , ' at ', data['messages'][hist_list[i]]["date"][11:] )
        
def consistency_interpreter(arr):  
    temp=[]
    for i in range (len(arr)):  
        if arr[i]<=1: temp.append(0)
        if arr[i]==2: temp.append(20)
        if arr[i]==3: temp.append(30)
        if arr[i]==4: temp.append(40)
        if arr[i]==5: temp.append(50)
        if arr[i]==6: temp.append(60)
        if arr[i]==7: temp.append(70)
        if arr[i]>7:  temp.append(80)
            
    return temp

def single_user_interaction_interpreter(telegram_name,reply_to_id):  #assumes access to message-id-to-telegram-name dict
    if reply_to_id not in set_of_good_message_ids_to_be_replied_to:
        return False
    
    reply_to_name= message_id_to_telegram_name_dict[reply_to_id]
    #if reply_to_name not in ['ArasH', 'arash', telegram_name]:  #no reply to teaher or themselves
    if reply_to_name not in [telegram_name]:  #no reply to themselves
        return True
    
    else:
        return False

def remove_return (list,ele):
    temp = copy.copy(list)
    temp.remove(ele)
    return temp    
    
    
    
def replace (arr, target, new):
    for i in range (len(arr)):
        if arr[i]==target:
            arr[i]=new
    return arr


def _draw_as_table(df, pagesize):
    alternating_colors = [['white'] * len(df.columns), ['aquamarine'] * len(df.columns)] * len(df)
    alternating_colors = alternating_colors[:len(df)]
    #fig, ax = plt.subplots(figsize=pagesize)
    fig, ax = plt.subplots(figsize=(11.69,8.27)) 
    ax.axis('tight')
    ax.axis('off')

    the_table = ax.table(cellText=df.values,
                        rowLabels=df.index,
                        colLabels=df.columns,
                        rowColours=['mediumspringgreen']*len(df),
                        colColours=['mediumspringgreen']*len(df.columns),
                        cellColours=alternating_colors,
                        loc='center')
    return fig
  

def dataframe_to_pdf(df, filename, numpages=(1, 1), pagesize=(11, 8.5)):
  with PdfPages(filename) as pdf:
    nh, nv = numpages
    rows_per_page = len(df) // nh
    cols_per_page = len(df.columns) // nv
    for i in range(0, nh):
        for j in range(0, nv):
            page = df.iloc[(i*rows_per_page):min((i+1)*rows_per_page, len(df)),
                           (j*cols_per_page):min((j+1)*cols_per_page, len(df.columns))]
            fig = _draw_as_table(page, pagesize)
            if nh > 1 or nv > 1:
                # Add a part/page number at bottom-center of page
                fig.text(0.5, 0.5/pagesize[0],
                         "Part-{}x{}: Page-{}".format(i+1, j+1, i*nv + j + 1),
                         ha='center', fontsize=8)
            pdf.savefig(fig, bbox_inches='tight')
            
            plt.close()
            
            
def append_to_excel(fpath, df, sheet_name):
    with pd.ExcelWriter(fpath, engine="openpyxl", mode="a") as f:
        df.to_excel(f, sheet_name=sheet_name)
        
def divide_list_by_four(array):
    
    list_to_be_returned = []
    
    num_of_sub_arrays = math.ceil(len(array)/4) 
    start_pointer = 0
    for i in range(0,num_of_sub_arrays):
        if i==num_of_sub_arrays-1:
            sub_array = array[start_pointer:]
        else:
            sub_array = array[start_pointer:start_pointer+4]
        start_pointer = start_pointer+4
        list_to_be_returned.append(sub_array)
        
    return list_to_be_returned
```
